## Предобработка данных

### Работа с пропущенными значениями

Для начала с помощью функции `isnull` и `sum` проверили количество пропусков по столбцам.

```python
print(sales.isnull().sum())
```

Пропуски есть как в данных, заполняемых пользователями, так и в данных, полученых автоматически на основе картографических данных. Необходимо последовательно изучить данные и понять что это за пропуски и что с ними делать.

Бросается в глаза 49 пропусков в населенном пункте **locality_name**. Он является одним из ключевых параметров нашего анализа и пропуски в нем не допустимы.

```python
locality_name_null = len(sales[sales['locality_name'].isnull()])/len(sales)
print('Доля пропусков по параметру "населенный пункт": {:.2%}'.format(locality_name_null))
```

Можно было бы определить какой населенный пункт пропущен, используя параметры, которые рассчитываются автоматически, например "расстояние до аэропорта". Но это не точный и трудозатратный метод, а так как доля пропусков в населенном пункте всего 0,21%, то разумнее удалить данные строки.

```python
print('Количество строк до удаления: ',len(sales))

sales.dropna(subset =['locality_name'], inplace=True)

print('Количество строк после удаления',len(sales))
```

По параметрам **apartment** и **balcony** можем предположить, что пропущенные значения связаны с их отсутствием у объекта недвижимости и при заполнении данных продавцы просто пропускли данные параметры. Заменили все пропуски в **apartment** на **False**, а в **balcony** на 0.

```python
sales['apartment'] = sales['apartment'].fillna(False)
print('Количество пропусков в параметры "Апартаменты": ', sales['apartment'].isnull().sum())

sales['balcony'] = sales['balcony'].fillna(0)
print('Количество пропусков в параметры "Число балконов": ', sales['balcony'].isnull().sum())
```

Далее посмотрели описание переменной "высота потолков в метрах" **ceiling_height** с помощью метода `describe`.

```python
sales['ceiling_height'].describe().to_frame()
```

Видим, что минимальное значение равно 1 метру, а максимальное - 100. Явно, что с этими данными что-то не так.  
На сегодняшний день в современных постройках минимальная высота потолка по нормам составляет 2,5 метра. Но на рынке мы можем встретить квартиры старого типа, а также частные дома. 
Также согласно современным тенденциям, на рынке встречаются квартиры с высокими потолками, предполагающими постройку второго этажа. Как правила высота таких потолков составляет не более 6 метров. 
Поэтому примем, что высота потолков находится в диапозоне от 2 до 6 метров.
Вывели список квартир с высотой потолка меньше 2 метров включительно и больше 6 метров.

```python
sales[(sales['ceiling_height']<=2) | (sales['ceiling_height']>6)].sort_values(by='ceiling_height')
```

Так как высота потолков у нас лежит в ограниченном диапазоне, все ошибочные данные заменим на среднее значение `mean`. В данных также есть значения типа 27 метров. Можно предположить, что при внесение данных, пользователь внес лишний ноль. Так как таких значений не много, заменим их на среднее.

```python
ceiling_height_mean = (
    sales[(sales['ceiling_height']>2) & (sales['ceiling_height']<6)]['ceiling_height'].mean())
print('Среднее значение высоты потолков: {:.2f}'.format(ceiling_height_mean))
```

В заданном диапазоне высоты потолков наше среднее значение уменьшилось.

Заменили все ошибки и заполнили пропуски на среднее значение.


```python
sales.loc[((sales.ceiling_height<=2) | (sales.ceiling_height>6)),
          'ceiling_height'] = ceiling_height_mean
sales['ceiling_height'] = sales['ceiling_height'].fillna(ceiling_height_mean)
print('Количество пропущенных значений:',sales['ceiling_height'].isnull().sum())
```

Проверили изменение с помощью метода `describe`.

```python
sales['ceiling_height'].describe().to_frame()
```

Все заменилось корректно. Посмотрели пропуски в параметре `floors_total`.

```python
floors_total_null = len(sales[sales['floors_total'].isnull()])/len(sales)
print('Доля пропусков по параметру "Всего этажей в доме": {:.2%}'.format(floors_total_null))
sales['floors_total'].describe().to_frame()
```

Видим, что минимальное количество этажей в доме - 1, а максимальное - 60 этажей.

Так как доля пропусков составляет всего 0,36%, то заменим все пропуски на среднее значение, округлив его до 10. При этом, если квартира находится на этаже (параметр **floor** ), который выше среднего значения, то пропуск заменим на значение из параметра floor.

```python
#выводит общее количество пропусков и количество пропусков, где параметр 'floor' больше 10
print('Общее количество пропусков в параметре "Общее количество этажей":',
      sales['floors_total'].isnull().sum())
print('Количество пропусков, где квартира находится выше 10 этажа:',
      len(sales[(sales['floors_total'].isnull())&(sales['floor'] > 10)]))
```

Заменили все пропуски в общем количестве этажей в доме, где фактический этаж расположения квартиры меньше или равен 10.

```python
sales.loc[((sales.floors_total.isnull()) & (sales.floor<=10)),
          'floors_total'] = 10

print('Общее количество пропусков в параметре "Общее количество этажей":',
      sales['floors_total'].isnull().sum())
```

У нас остались только 34 пропущенных значения, где фактический этаж расположения квартиры выше 10. Заменили и их.

```python
sales.loc[((sales.floors_total.isnull()) & (sales.floor>10)),
          'floors_total'] = sales.loc[((sales.floors_total.isnull()) & (sales.floor>10)),
          'floor'] 
print('Общее количество пропусков в параметре "Общее количество этажей":',
      sales['floors_total'].isnull().sum())
```

Пропусков в параметре **floors_total** больше нет. Перешли к пропускам в параметрах, относящихся к площади недвижимости: **total_area, living_area и kitchen_area**.

Рассмотрим, как соотносится жилая площадь и площадь кухни с общей площадью квартиры. Соотношение посмотрим в разрезе параметра ***количество комнат rooms**. Для начала посмотрели, что находится в данных параметра `rooms` с помощью метода `describe`.

```python
sales['rooms'].describe().to_frame()
```

В данных есть нулевые значения, возможно это связано с типом планировки: "студия" и "свободная планировка". Рассмотрели их с учетом группировки по данным двум параметрам.

```python
print('Количество 0 значений в параметре "rooms":', len(sales.query('rooms == 0')))
sales.query('rooms == 0').groupby(['studio','open_plan'])['rooms'].count().to_frame()
```

Как и предполагали, все пропуски относятся к типу планировки жилья. Так как доля их невелика, мы можем заменить нулевые значения на средние, не исказив при этом данные. Посмотрели, какая площадь квартир с 0 количеством комнат.

```python
sales.query('rooms == 0')['total_area'].describe().to_frame()
```

Минимальное значение 15 кв.м., а вот максимальное 371 кв.м. Отсортировали таблицу и посмотрели последние 15 строк.

```python
sales.query('rooms == 0').sort_values('total_area').tail(15)
```

Есть квартиры с большой  площадью, это и 58 кв.м, и 98кв.м. и даже 371 кв.м. Рассмотрели, какая средняя площадь в разрезе количества комнат, по которым есть данные. Так как в данных есть квартиры с большой площадью, то в качестве среднего значения использовали медиану.

```python
sales.query('rooms != 0').groupby('rooms')['total_area'].median().to_frame()
```

Для количества комнат с 1 по 4 площадь увеличивается, а вот дальше идет непредсказуемый разброс. Заменили нулевые значения исходя из следующего: площадь до 40 кв.м. включительно - 1 комната, 40-60 кв.м. - 2 комнаты, 60-90 кв.м. - 3 комнаты, 90-120 кв.м. - 4 комнаты. А для квартиры в 300 кв.м. рассчитали свое среднее значение.

```python
#заменяет 0 значение на 1 для квартир, площадью до 40 кв.м. включительно
sales.loc[((sales['rooms'] == 0) & (sales.total_area <= 40)),'rooms'] = 1

#заменяет 0 значение на 2 для квартир, площадью до 40-60 кв.м. включительно
sales.loc[((sales['rooms'] == 0) & (sales.total_area > 40)&
           (sales.total_area <= 60)),'rooms'] = 2

#заменяет 0 значение на 2 для квартир, площадью до 60-90 кв.м. включительно
sales.loc[((sales['rooms'] == 0) & (sales.total_area > 60)&
           (sales.total_area <= 90)),'rooms'] = 3

#заменяет 0 значение на 2 для квартир, площадью до 90-120 кв.м. включительно
sales.loc[((sales['rooms'] == 0) & (sales.total_area > 90)&
           (sales.total_area <= 120)),'rooms'] = 4

sales.query('rooms == 0')
```

С нулевым значением осталась только квартира площадью 371 кв.м. Рассчитали среднее значение с использованием медианы и заменили на него нулевое значение.

```python
room_300_median = sales.query('total_area > 300')['rooms'].median()
sales.loc[((sales['rooms'] == 0) & (sales.total_area == 371)),'rooms'] = room_300_median
print('Минимальное количество комнат:',sales['rooms'].min())
```

Теперь у нас есть все данные в параметре **rooms**. Вывели данные, в которых присутствуют все значения площади, и сохранили их в **area**.

```python
area = sales[(~sales['kitchen_area'].isnull())&
             (~sales['living_area'].isnull())][['locality_name',
                                                'rooms',
                                                'total_area',
                                                'living_area',
                                                'kitchen_area']].reset_index()

part_area_full = len(area)/len(sales)
print('Доля заполненных данных по площади: {:.1%}'.format(part_area_full))
display(area)
```

Всего у нас 20943 строки с полными данными по площади, это почти 89% всех данных. Таким образом, получившиеся в результате средние соотношения можем смело применить к недостающим данным, это не приведет к их искажению.
Рассчитали соотношение площадей к общей площади.

```python
#добавляет сумму жилой площади и кухни в столбец 'kitchen_living'
area['kitchen_living'] = area['living_area'] + area['kitchen_area']

#добавляет в столбец 'share_kitchen_living_in_total' какую долю в общей площади квартиры 
#составляет сумма жилой площади и кухни вместе
area['share_kitchen_living_in_total'] = (area['kitchen_living'] / area['total_area'])

#добавляет в столбец 'share_kitchen_in_total' какую долю в общей площади квартиры 
#занимает кухня 
area['share_kitchen_in_total'] = (area['kitchen_area'] / area['total_area'])

#добавляет в столбец 'share_living_in_total' какую долю в общей площади квартиры 
#занимает жилая площадь
area['share_living_in_total'] = (area['living_area'] / area['total_area'])

area.head()
```

С помощью метода `describe` посмотрели на получившиеся данные.

Принимаем, что в населенных пунктах, у которых значение delta равно 0 (т.е. во всех объявлениях населенного пункта пропушен данный параметр) нет поблизости аэропорта, такие значения заполним значением, значительно превышающим текущее максимальное, возьмем 120000.
```python
#общее количество объявлений в разрезе городов
locality_name_total = sales.pivot_table(index='locality_name',
                                       values='last_price',
                                       aggfunc='count')
locality_name_total.columns = ['count_total']

#количество объявлений, в разрезе городов, там где есть данные в параметре airports_nearest
airports_locality_name_not_null = sales[~sales['airports_nearest'].isnull()].pivot_table(
    index='locality_name',
    values='airports_nearest',
    aggfunc=['min','median','mean','max'])

airports_locality_name_not_null.columns = ['airports_nearest_min',
                                 'airports_nearest_median',
                                 'airports_nearest_mean', 
                                 'airports_nearest_max'] 

#количество объявлений, в которых не указан параметр airports_nearest
locality_name_airpots = sales[sales['airports_nearest'].isnull()].pivot_table(
    index='locality_name',
    values='last_price',
    aggfunc='count')

locality_name_airpots.columns = ['count_null']


#объединяет все три таблицы в одну locality_name_airpots
locality_name_airpots = locality_name_airpots.merge(locality_name_total,
                                                 on='locality_name',
                                                 how='left')

locality_name_airpots = locality_name_airpots.merge(airports_locality_name_not_null,
                                                 on='locality_name',
                                               how='left')

#добавляет параметр delta в конечную таблицу locality_name_airpots
locality_name_airpots['delta'] = (locality_name_airpots['count_null'] - 
                                locality_name_airpots['count_total'])

display(locality_name_airpots)
```

```python
sales['price_square_meter'] = (sales['last_price'] / sales['total_area'])
sales.head()
```

Далее добавили день недели, месяц и год публикации объявления.

```python
sales['first_day_exposition_weekday'] = pd.DatetimeIndex(sales['first_day_exposition']).weekday
sales['first_day_exposition_month'] = pd.DatetimeIndex(sales['first_day_exposition']).month
sales['first_day_exposition_year'] = pd.DatetimeIndex(sales['first_day_exposition']).year
sales.head()
```

Добавили параметр **floor_type** тип квартиры: первый этаж, последний этаж и другой.

```python
def floor_type(row):
    #Функция floor_type принимает на вход строку данных из DataFrame.
    #Далее в зависимости от этажа выводит нужное значение: 1 этаж - первый, 
    #последний этаж - последний, для всех остальных - другой.
    
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
sales['floor_type'] = sales.apply(floor_type, axis=1)
sales.head()
```

Посмотрим данные в разрезе городов.

```python
def time_too_fast(row):
    
    #функция time_too_fast принимает на входе ряд данных и в зависимости от значения параметра days_exposition
    #выводит True или False
    
    time = row['days_exposition']
    
    if time < 30:
        return True
    return False

#добавляет столбец too_fast
sales['too_fast'] = sales.apply(time_too_fast, axis=1)

display(sales.head())
```

Далее рассмотрели зависимость цены от того, на каком этаже расположена квартира: первом, последнем или другом.

Расчитали срденее значение как медиану для каждого типа этажа для разных годов и построили график.

```python
floor_price = sales_good.pivot_table(index='first_day_exposition_year' ,
                                     columns='floor_type',
                                     values='last_price',
                                     aggfunc='median')

floor_price_median_total = sales_good.groupby('first_day_exposition_year')['last_price'].median()

floor_price = floor_price.join(floor_price_median_total)

floor_price.columns = ['another_median', 'first_median', 'last_median', 'total_median']

floor_price['deviation_first_to_total'] = ((floor_price['first_median']/floor_price['total_median'] - 1) *100)
floor_price['deviation_last_to_total'] = ((floor_price['last_median']/floor_price['total_median'] - 1) *100)
floor_price['deviation_another_to_total'] = ((floor_price['another_median']/floor_price['total_median'] - 1) *100)
floor_price
```

Получили, что самая низкая стоимость у квартир, расположенных на первом этаже. Самая высокая стоимость у квартир, которые не находятся ни на первом, ни на последнем этажах. При этом стоимость квартир на первом и последнем этажах ниже среднерыночной стоимости, а все остальные этажи выше среднерыночной стоимсоти.

Построили график.

```python
#отображает 1000000 руб. как 1 млн.руб.
def lblFormat(n, pos):
    return str(int(n / 1e6))
lblFormatter = FuncFormatter(lblFormat)

floor_price.plot(y=['another_median','first_median','last_median','total_median'],
                 style='o-',
                 ylim=(2000000,8000000),
                 label=['Цена за другой этаж','Цена за первый этаж','Цена за последний этаж','Среднерыночная цена'],
                 grid=True)

#добавляет названия осей
plt.xlabel('Год', fontsize=12)
plt.ylabel('Стоимость, млн.руб.', fontsize=12)
plt.title('Средняя стоимость квартиры в зависимости от этажа в динамике по годам')
ax.legend(['другой этаж','первый этаж','последний этаж','среднерыночная'])

#изменяет отображение значений оси y как 1 млн.руб.
ax = plt.gca()
ax.yaxis.set_major_formatter(lblFormatter)
plt.show()
```

В кризисный 2014 год средняя стоимочть квартир была выше, чем в последующие года. В 2014-2015 годах минимальная стоимость была у квартир, расположенных на последнем этаже. С 2017 года наблюдаем одинаковую тенденцию для стоимости квартир в зависимости от этажа расположения.
Посмотрели среднее отклонение цен от среднерыночной цена за 2017-2019 года.

```python
print('Цена квартир на первом этаже ниже среднерыночной в среднем на {:.1f} процентов'.format(
      floor_price.query('index in ["2017","2018","2019"]')['deviation_first_to_total'].mean()))

print('Цена квартир на последнем этаже ниже среднерыночной в среднем на {:.1f} процентов'.format(
      floor_price.query('index in ["2017","2018","2019"]')['deviation_last_to_total'].mean()))

print('Цена квартир на других этажах выше среднерыночной в среднем на {:.1f} процентов'.format(
      floor_price.query('index in ["2017","2018","2019"]')['deviation_another_to_total'].mean()))
```

Таким образом, можем сказать, что взаимосвязь стоимости квартир и этажа ее расположения присутствует. Зная этаж квартиры и текущую среднюю рыночную стоимость, мы можем предсказать стоимость квартиры.

```python
#создает срез по квартирам в центре Санкт-Петербурга по типу этажа и цене
spb_center_floor = spb_center.pivot_table(index='floor_type',values='last_price',aggfunc='median')

#создает срез по квартирам во всем Санкт-Петербурге  по типу этажа  и цене
spb_floor = sales_good.query('locality_name == "Санкт-Петербург"').pivot_table(index='floor_type',values='last_price',aggfunc='median')

#объединяет две таблицы
spb_center_floor = spb_center_floor.merge(spb_floor, how='outer',right_index=True,left_index=True, suffixes=('_center','_total'))

#добавляет соотношение средней стоимости в зависимости от этажа к средней по всей выборке
spb_center_floor['deviation_center'] = ((spb_center_floor['last_price_center']/
                                         spb_center['last_price'].median() - 1) *100)

spb_center_floor['deviation_total'] = ((spb_center_floor['last_price_total']/
                                         sales_good.query('locality_name == "Санкт-Петербург"')['last_price'].median() - 1) *100)

#выводит таблицу
display(spb_center_floor)

#строит график
spb_center_floor.plot(y=['last_price_center', 'last_price_total'],
                      kind='bar',
                      ylim=(0,9000000),
                      grid=True,
                      label=['Центр','Весь город'],
                      title = 'Средняя стоимость квартиры в зависимости от этажа'
                      )

#добавляет названия осей
plt.xlabel('Этаж', fontsize=12)
plt.ylabel('Стоимость, млн.руб.', fontsize=12)

#изменяет отображение значений оси y как 1.0 млн.руб.
ax = plt.gca()
ax.yaxis.set_major_formatter(lblFormatter_float)
plt.show()
```

    - Для квартир в центре в будние дни цена размещения изменяется не сильно, а вот в выходные наблюдается сильный рост. По всему городу зависимости от дней недели не наблюдается.
