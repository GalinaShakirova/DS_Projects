Посмотрим на количество пропущенных значений.

```python
games[games['name'].isnull()]
```

Видим, что значения также пропущены и в параметре **genre**, эти данные не сможем использовать в анализе, поэтому удалим их.

```python
#вывод общего числа строк до удаления
print('Количество строк до удаления: ',len(games['name']))

#удаление строк с пропуском в названии игры 'name'
games.drop(games[games.name.isnull()].index, inplace=True)

#вывод общего числа строк после удаления пропусков в поле 'name'
print('Количество строк после удаления',len(games))
games.isnull().sum().to_frame()
```

У нас осталось 4 параметра с пропущенными значениями: год выпуска **year_of_release**, оценка критиков **critic_score**, оценка пользователей **user_score** и рейтинг от ассоциации ESRB **rating**. Рассмотрим параметр год выпуска.

```python
print('Доля пропусков в параметре "Год выпуска": {:.1%}'.format(
    len(games[games['year_of_release'].isnull()])/len(games)))
games[games['year_of_release'].isnull()].head()
```

Доля пропущенных значений всего 1,6%. В данных есть игры, которые выпускались в разные годы для разных платформ. Посмотрим, какую долю они составляют.


```python
#формирует сводную таблицу с индексами название игры и годом выпуска
games_group_name_year = games.pivot_table(
    index=['name','year_of_release'],aggfunc={'platform':'count'}).reset_index()

#формирует сводную таблицу с индексом названия игры и количеством разных лет выпуска
name_year = games_group_name_year.pivot_table(
    index='name',aggfunc={'year_of_release':'count','platform':'sum'})

print('Количество игр, выпущенных в разные года для разных платформ: {}, их доля - {:.0%}'.format(
    len(name_year.query('year_of_release > 1')), len(name_year.query('year_of_release > 1'))/len(games)))

#сортирует таблицу
name_year.sort_values(by='year_of_release', ascending=False).head(10)
```

Доля таких игр составляет всего 4%. В связи с этим для всех игр, которые по известным данным выпускались в один год на разных платформах, заменим пропуски на этот год. Там где игры выпускались в разные годы использовать какой-либо год для замены пропусков некорректно. Поэтому все остальные пропуски заменим на 0.

Сформируем таблицу, с данными о играх, выходивших в один год на разных приставках, и годом выпуска.

```python
#формирует сводную таблицу с индексами название игры и годом выпуска
games_group_name_year_first = games_group_name_year.pivot_table(
    index='name',aggfunc={'year_of_release':'first'})

#переименовывает столбец
name_year = name_year.rename(columns={'year_of_release':'count_year'})

#формирует таблицу с играми, выпускавшимися в один год на разных платформах
name_year = name_year.query('count_year == 1')

#объединяет две таблицы
name_year = name_year.merge(games_group_name_year_first, on='name', how='left')
name_year.head()
```

Заполним все пропущенные значения на 0, а также посмотрим информацию о игре, по которой будем проверять корректность заполнения данных.

```python
print('Количество пропусков до заполнения:', games['year_of_release'].isnull().sum())

#заменяет пропуски на 0
games['year_of_release'] = games['year_of_release'].fillna(0)

print('Количество пропусков после заполнения:', games['year_of_release'].isnull().sum())
print('Количество значений с 0 оценкой критиков:', len(games.query('year_of_release == 0')))

#выводит информацию по игре, по которой будем проверять заполнение пропусков на среднее значение
games.query('name == "LEGO Harry Potter: Years 5-7"')
```

Заполним пропущенные значения там, где это возможно. В остальных случаях оставим значение 0.

```python
def fill_year_of_release(row, data):
    """Принимает строку и DataFrame."""
    name = row['name']
    year_of_release = row['year_of_release']
    
    """
    Если в 'year_of_release' текущей таблицы значение 0 и название игры есть в значениях
    таблицы data, то возвращает значение new_value из таблицы data. 
    Если нет - возвращает текущее значение параметра в исходной таблице.
    """
    if (name in data.index) and (year_of_release == 0):
        
        new_value = data.loc[name,'year_of_release']
        return new_value
    
    return year_of_release

#заменяет 0 на год выпуска
games['year_of_release'] = games.apply(lambda row: fill_year_of_release(row, name_year), axis=1)

#выводит информацию для проверки
print('Количество пропусков после заполнения:', games['year_of_release'].isnull().sum())
print('Количество значений с 0 годом выпуска:', len(games.query('year_of_release == 0')))
games.query('name == "LEGO Harry Potter: Years 5-7"')
```

Пропуски заполнили корректно. Теперь перейдем к пропускам в параметрах **critic_score, user_score и rating**.

```python
for column in ['critic_score', 'user_score','rating']:
    print('Доля пропущенных значений в параметре "'+column+'": {:.1%}'.format(
        games[column].isnull().sum()/len(games)))
```

Пропущенных значений долстаточно много. Посмотрим, есть ли взаимосвязь пропусков в параметрах с годом выхода игры. Для этого заменим пропуски в параметрах **user_score и rating** на "нет данных" и присвоим им значение 0, а там где есть данные - 1. Таким образом мы сможем понять сколько игр за определенный год имело оценку или рейтинг.

```python
#заменяет пропуски
games['user_score'] = games['user_score'].fillna('нет данных')
games['rating'] = games['rating'].fillna('нет данных')

#где есть NaN или "нет данных" в параметрах ставит 0, а где есть значение - 1
games['have_critic_score'] = (games['critic_score'] >= 0)*1
games['have_user_score'] = (games['user_score'] != 'нет данных')*1
games['have_rating'] = (games['rating'] != 'нет данных')*1

#формирует сводную таблицу
games.pivot_table(index='year_of_release', 
                  aggfunc={'na_sales':'count','have_critic_score':'sum',
                          'have_user_score':'sum','have_rating':'sum'})
```

Исходя из таблицы видим, что до 2000 года всех трех параметров почти не было, что вполне объяснимо. Интернет получил широкое распростронение в конце 20 века, именно по этой причине отсутствует информация. Если и есть единичные данные, то скорее всего они были поставлены поклонниками после 2000г. Кроме того, до того, как компьютеры стали доступны для населения и появились почти в каждом доме, а так же до широкого расспространия интернета, на рынке присутствовали игровые приставки к телевизору и консоли, а также игры к ним. Широкое расспространиение получили PS (известная как PlayStation), NES (известная как Dendy), GameBoy, Super NES и Nintendo 64. Вся информация об играх, их оценках критиками и рейтингах выходили в журналах.

```python
#формирует таблицу с количеством игровых приставок, для которых была выпущена игра
count_plat_for_game = games.groupby('name')['platform'].count().to_frame().reset_index()

#переименовывает столбец и выводит таблицу
count_plat_for_game = count_plat_for_game.rename(columns={'platform': 'count_plat_for_game'})
count_plat_for_game.head()
```

Добавим количество платформ, для которых выпускались игры, в таблицу **games**.

```python
#формирует сводную таблицу в разрезе игр и оценок критиков
critic_score_full = games[((~games['critic_score'].isnull()) &
                    (games['count_plat_for_game']>1))].pivot_table(
    index='name',values='critic_score', aggfunc=['count','min','mean','max'])

#переименовывает колонки
critic_score_full.columns = ['count','min_critic_score','mean_critic_score','max_critic_score']

#добавляет параметр: на сколько процентов максимальная оценка превышает минимальную
critic_score_full['per_max_to_min'] = (critic_score_full['max_critic_score'] /
                                       critic_score_full['min_critic_score'] - 1)

#сортирует таблицу
critic_score_full.sort_values(by='per_max_to_min', ascending=False)
```

Посмотрим какую долю занимают игры, которые выпускаются для 2 и более платформ, имеющих оценку критиков, в общем количестве игр, выпущенных для 2 и более платформ.

```python
print('Доля игр, выпускавшихся для 2 и более платформ, для которых есть оценка критиков: {:.0%}'.format(
    len(critic_score_full)/len(games.query('count_plat_for_game > 1')['name'].unique())))
```

```python
#Строит график
critic_score_full.query('count > 1').plot(kind='hist',
                                   y='per_max_to_min',
                                   range=(0,0.6),
                                   grid=True,
                                   figsize=(9,6),
                                   bins=70,
                                  legend=False)

#Выводит формат для гистограммы
plt.title('Распределение отклонения максимальной оценки критиков от минимальной')
plt.xlabel('Процент отклонения')
plt.ylabel('Количество игр, штук')
plt.show()
```

Исходя из гистограммы видим, что для большинства игр максимальная и минимальная оценка отличаются  на 10% и менее. А все остальные пропуски для игр до 2000 и после заменим на 0. Причина пропусков после 2000 года могут быть обусловлены тем, что, например, данные не были внесены.
```python
print('Количество пропусков до заполнения:', games['critic_score'].isnull().sum())

#заменяет пропуски на 0
games['critic_score'] = games['critic_score'].fillna(0)
print('Количество пропусков после заполнения:', games['critic_score'].isnull().sum())
print('Количество значений с 0 оценкой критиков:', len(games.query('critic_score == 0')))

#выводит информацию по игре, по которой будем проверять заполнение пропусков на среднее значение
games.query('name == "Call of Duty: Advanced Warfare"')
```

Заполним пропущенные значения на среднее значение.

```python
def fill_critic_score(row, data):
    """Принимает строку и DataFrame."""
    name = row['name']
    year_of_release = row['year_of_release']
    critic_score = row['critic_score']
    
    """
    Если в параметре 'critic_score' текущей таблицы значение 0 и название игры есть в индексе
    таблицы data, то возвращает среднюю оценку критиков. Если нет - возвращает текущее значение
    параметра в исходной таблице.
    """
    if (name in data.index) and (critic_score == 0) and (year_of_release >= 2012):
        mean_critic_score = data.loc[name,'mean_critic_score']
        return mean_critic_score
    
    return critic_score

games['critic_score'] = games.apply(lambda row: fill_critic_score(row, critic_score_full), 1)
```

Проверим изменения.

```python
print('Количество пропусков после заполнения:', games['critic_score'].isnull().sum())
print('Количество значений с 0 оценкой критиков:', len(games.query('critic_score == 0')))
print('Средняя оценка критиков для игры "Call of Duty: Advanced Warfare": {:.2f}'.format( 
      critic_score_full.loc['Call of Duty: Advanced Warfare','mean_critic_score']))
#выаодит информацию по игре, по которой будем проверять заполнение пропусков на среднее значение
games.query('name == "Call of Duty: Advanced Warfare"')
```

Сделаем аналогичные действия для оценок пользователей. В параметре "Оценка пользователей" у нас также есть значение **"tbd"**, что расшифровывается как **'To be determined'** - это значит компания объявила о разработке игры, но конечная даты выхода еще не определена. Когда tbd появляется в параметре оценки пользователей, это означает, что игра добавлена в базу, но оценок по ней быть не может. Когда игра выходит, то, как правило, tbd заменяется на оценку пользователей. Наличие tbd может свидетельствовать либо о технической ошибке, либо о том, что пользователи не проставили оценку по каким-либо причинам (игра была не интересно и другое), или же игра так и не была выпущена.

```python
#формирует сводную таблицу в разрезе игр и оценок критиков
user_score_not_null = games.query('user_score != "нет данных" and user_score != "tbd"').copy()
user_score_not_null['user_score'] = user_score_not_null['user_score'].astype('float')

user_score_full = user_score_not_null.query('count_plat_for_game >1').pivot_table(
    index='name',values='user_score', aggfunc=['count','min','mean','max'])


#переименовывает колонки
user_score_full.columns = ['count','min_score','mean_score','max_score']

#добавляет параметр: на сколько процентов максимальная оценка превышает минимальную
user_score_full['per_max_to_min'] = (user_score_full['max_score'] /
                                user_score_full['min_score'] - 1)

#сортирует таблицу
user_score_full.sort_values(by='per_max_to_min', ascending=False)
```

```python
print('Количество значений "нет данных" в оценки пользователей:', len(games.query('user_score == "нет данных"')))
print('Количество значений "tbd" в оценки пользователей:', len(games.query('user_score == "tbd"')))

#выводит информацию по игре, по которой будем проверять заполнение пропусков на среднее значение
games.query('name == "Ride"')
```

Заполним пропущенные значения на среднее значение.

```python
def fill_user_score(row, data):
    """Принимает строку и DataFrame."""
    name = row['name']
    year_of_release = row['year_of_release']
    user_score = row['user_score']
    
    """
    Если в параметре 'user_score' текущей таблицы значение "нет данных" или "tbd" и:
      - название игры есть в индексе таблицы data, то возвращает среднюю оценку критиков,
      - название игры нет в индексе таблицы data, то возвращает 0.
    Во всех остальных случаях - возвращает текущее значение параметра в исходной таблице.
    """
    if (name in data.index) and ((user_score == "нет данных") or (user_score == "tbd")) and (year_of_release >= 2000):
        mean_user_score = data.loc[name,'mean_score']
        return mean_user_score
    elif (user_score == "нет данных") or (user_score == "tbd"):
        return 0
    
    return user_score

games['user_score'] = games.apply(lambda row: fill_user_score(row, user_score_full), 1)
```

```python
#формирует сводную таблицу
year_genre = games.query('year_of_release > 2005').pivot_table(index='year_of_release',
                                                            columns='genre', 
                                                            aggfunc={'total_sales':'sum'})

#заменяет суммы продаж по каждой платформе за год доле в глобальных продажах за каждый год
for column in year_genre.columns:
    for year in year_genre.index:
        
        year_genre.loc[year,column] = (year_genre.loc[year,column] /
                                       games.query('year_of_release==@year')['total_sales'].sum())
#переименовывает колонки
year_genre.columns = ['Action','Adventure','Fighting','Misc','Platform','Puzzle','Racing','Role-Playing',
                      'Shooter','Simulation','Sports','Strategy']

#переворачивает и выводит таблицу
year_genre.T
```

```python
#добавляет парметры top_4 - доля основных 4-х жанров, и other - суммарная доля остальных платформ
year_genre['top_4'] = (year_genre['Action'] + year_genre['Role-Playing'] + 
                       year_genre['Shooter'] + year_genre['Sports'])

year_genre['other'] = (year_genre['Adventure'] + year_genre['Fighting'] + year_genre['Misc'] + 
                       year_genre['Platform'] + year_genre['Puzzle'] + year_genre['Racing'] +
                      year_genre['Simulation'] + year_genre['Strategy'])

#выводит таблицу
year_genre[['top_4','other']].T
```

До 2010 года доля Action, Shooter, Role-Playing и Sports в общем объеме составляла в среднем 56%, а с 2011 до 2016 года ее доля выросла с 64% до 85%.

Рассмортим продажи игр в разрезе платформ для каждого региона отдельно.

```python
#формирует сводную таблицу
games_region = games_net.copy()

#формирует сводную таблицу
top_platform_region = games_region.pivot_table(index='platform', 
                      values=['na_sales','eu_sales','jp_sales','total_sales'],
                      aggfunc='sum', 
                      fill_value=0).sort_values(by='na_sales', ascending=False).reset_index()


#убирает нулевые продажи
top_platform_region.drop(top_platform_region[top_platform_region['total_sales']==0].index, inplace=True)

def plot_region(data,parametr,title,xlabel): 
    
    #строит график
    data.plot(y=['na_sales', 'eu_sales','jp_sales'],
                             x = parametr,
                             kind='bar',
                             figsize=(9,6),
                             grid=True,
                             label=['Северная Алмерика','Европа','Япония'])


    #задает формат для гистограммы
    plt.title('Распределение продаж игр в разрезе '+title+' и региона', fontsize=15)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel('Продажи, млн. копий', fontsize=12)
    plt.show()
    
plot_region(top_platform_region,'platform','платформ','Платформа')
```

## Проверка гипотез

В ходе исследования выдвинули две гипотезы:
- средние пользовательские рейтинги платформ XOne и PC равны,
- средние пользовательские рейтинги жанров Action и Sports разные.

Проверим их. Критический уровень статистической значимости примем 5%.

### Проверка гипотезы о равенстве средних пользовательских рейтингов платформ XOne и PC.

Сформируем Н0 и Н1 гипотезы.

Гипотеза Н0 — средние пользовательские рейтинги платформ XOne и PC равны.


Гипотеза Н1 — средние пользовательские рейтинги платформ XOne и PC равны различаются.

Так как выборки у нас независимые, воспользуемся методом ttest_ind из модуля scipy.stats. 
Посмотрим, отличаются ли показатели дисперсий в рассматриваемом периоде для данных платформ.

```python
games_net.query('(platform == "PC" or platform == "XOne") and user_score > 0').pivot_table(
    index='platform', aggfunc={'user_score': np.var})
```

Дисперсии распределения пользовательских рейтингов отличаются не сильно, поэтому примем, что дисперсии двух совокупностей примерно равные.

Проверим нашу гипотезу.

```python
#Формирует массив оценок пользователей для платформы XOne
xone_user_score = games_net.query('platform == "XOne" and user_score > 0')['user_score']

#Формирует массив оценок пользователей для платформы PC
pc_user_score = games_net.query('platform == "PC" and user_score > 0')['user_score']

#Задает критический уровень статистической значимости
alpha = 0.05 

#проверяет гипотезу
results = st.ttest_ind(xone_user_score, pc_user_score)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Нулевую гипотезу отвергнуть не получилось. Можем предположить, что средние пользовательские рейтинги игр для платформ PC и XOne если и отличаются, то не значительно.

Посмотрим на диаграмму размаха данных платформ за анализируемый период.

```python
#формирует таблицу
subset = games_region.query('(platform == "PC" or platform == "XOne") and user_score > 0').copy()
subset['platform'] = subset['platform'].astype(object)

#выводит диаграмму размаха
sns.catplot(x="platform", 
            y='user_score', 
            kind="box", 
            data= subset, 
            orient='v',
            palette="Set3")

#задает форматирование
plt.title('Диаграмма размаха оценок пользователей для платформ за анализируемый период', fontsize=14)
plt.ylim(0,10)
plt.xlabel('Платформы', fontsize=13)
plt.ylabel('Оценки пользователей, балл', fontsize=13)
plt.grid(True)
plt.show()
```

Видим, что в анализируемом периоде платформы Xone и PC имеют одинаковые медианы и очень схожий обхват пользовательских оценок.

### Проверка гипотезы о различии средних пользовательских рейтингов жанров Action и Sports.

Сформируем Н0 и Н1 гипотезы.

Гипотеза Н0 — средние пользовательские рейтинги жанров Action и Sports равны.

Гипотеза Н1 — средние пользовательские рейтинги жанров Action и Sports различаются.

Так как выборки у нас независимые, воспользуемся методом ttest_ind из модуля scipy.stats. 
Посмотрим, отличаются ли показатели дисперсий в рассматриваемом периоде для данных платформ.

```python
games_net.query('(genre == "Action" or genre == "Sports") and user_score > 0').pivot_table(
    index='genre', aggfunc={'user_score': np.var})
```

Дисперсии распределения пользовательских рейтингов отличаются не сильно, поэтому примем, что дисперсии двух совокупностей примерно равные.

Проверим нашу гипотезу.

```python
#Формирует массив оценок пользователей для жанра Action
action_user_score = games_net.query('genre == "Action" and user_score > 0')['user_score']

#Формирует массив оценок пользователей для жанра Sports
sports_user_score = games_net.query('genre == "Sports" and user_score > 0')['user_score']

#Задает критический уровень статистической значимости
alpha = 0.05 

#проверяет гипотезу
results = st.ttest_ind(action_user_score, sports_user_score)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Нулевая гипотеза не подтвердилась. Таким образом у нас есть все основания полагать, что средние пользовательские рейтинги для жанров Action и Sports различаются.

Посмотрим на диаграмму размаха данных жанров за анализируемый период.

```python
#формирует таблицу
subset_genre = games_region.query('(genre == "Action" or genre == "Sports") and user_score > 0').copy()
subset_genre['genre'] = subset_genre['genre'].astype(object)

#выводит диаграмму размаха
sns.catplot(x="genre", 
            y='user_score', 
            kind="box", 
            data= subset_genre, 
            orient='v',
            palette="Set3")

#задает форматирование
plt.title('Диаграмма размаха оценок пользователей для платформ за анализируемый период', fontsize=15)
plt.ylim(0,10)
plt.xlabel('Платформы', fontsize=13)
plt.ylabel('Оценки пользователей, балл', fontsize=13)
plt.grid(True)
plt.show()
```

В нашей выборке данных медианы распределения пользовательских оценок отличаются, для жанра Action она выше. Также отличается охват пользовательских оценок.
