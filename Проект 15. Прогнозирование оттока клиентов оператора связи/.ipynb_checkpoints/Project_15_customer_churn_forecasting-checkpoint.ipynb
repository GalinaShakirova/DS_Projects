# Прогнозирование оттока клиентов оператора связи

**Заказчик:** оператор связи «Ниединогоразрыва.ком».
Оператор хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия.

**Исходные данные:** команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах. **Информация о договорах актуальна на 1 февраля 2020 года.**

Основные услуги оператора:
1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также оператор предоставляет дополнительные услуги, такие как:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Задача** - построить модель для прогноза оттока клиентов.

**Описание полей данных:**
- BeginDate – дата начала пользования услугами,
- EndDate – дата окончания пользования услугами,
- Type – тип оплаты: ежемесячный, годовой и тд,
- PaperlessBilling – безналичный расчет,
- PaymentMethod – способ оплаты,
- MonthlyCharges – ежемесячные траты на услуги,
- TotalCharges – всего потрачено денег на услуги
- Dependents – наличие иждивенцев
- Senior Citizen – наличие пенсионного статуса по возрасту
- Partner – наличие супруга(и)
- MultipleLines – наличие возможности ведения параллельных линий во время звонка

6. Посчитаем дополнительный параметр - количество подключенных услуг оператором.

Чтобы проверить корректность замены значений **Yes, No и Do not have** в параметрах с дополнительными услугами оператора, посмотрим на распределение долей до изменения.

```python
#список параметров дополнительных услуг оператора
columns_additional_service = ['online_security','online_backup','tech_support',
                              'device_protection','streaming_tv','streaming_movies']

for column in columns_additional_service:
    display(df[column].value_counts(normalize = True, dropna=False).to_frame())
```

Проведем предобработку данных.

```python
def active_month(row):
    '''
    Считает количество месяцев, в течении которого клиент пользуется услугами компании. 
    Если конечной даты нет - берет последнюю дату в выборке
    '''
    last_date = pd.to_datetime('2020-02-01')
    
    if row['end_date'] == 'No':
        return int(round((last_date - row['date_start'])/np.timedelta64(1, 'M'),0))
    else:
        return int(round((pd.to_datetime(row['end_date']) - row['date_start'])/np.timedelta64(1, 'M'),0))


def change_on_number(df, columns):
    '''Принимает строку DataFrame. Изменяет ответы No и NuN на 0, Yes - на 1'''
    df=df.copy()
    for column in columns:
        for row in range(len(df)):
            if df.loc[row, column] == 'Yes':
                df.loc[row, column] = 1
            elif df.loc[row, column] == 'No':
                df.loc[row, column] = 0
            else:
                df.loc[row, column]= 0
    return df

def change_type(df, columns, type_on):
    '''Изменяет тип данных на заданный'''
    df = df.copy()
    for col in columns:
        df[col] = df[col].astype(type_on)
    return df

def count_of_service(row):
    '''Считает количество подключенных услуг'''
    count_of_service = row[columns_additional_service].sum() + row['have_phone'] + row['have_internet']
    return count_of_service

    
    
#предобработка данных
df = (df
      .assign(
          #добавляет целевой признка
          active_cust = lambda x: x['end_date'].apply(lambda x: 0 if x == 'No' else 1),
          #переводит параметр даты заключения договора в формат даты
          date_start = lambda x: pd.to_datetime(x['begin_date']),
          #формирует параметр года заключения договора
          date_start_year = lambda x: x['date_start'].dt.year,
          #рассчитывает количество месяцев, в течении которых действует договор
          month_active = lambda x: x.apply(active_month, axis=1),
          #добавляет признак наличия стационарной телефонной связи в пакете услуг клиента
          have_phone = lambda x: x['multiple_lines'].apply(lambda x: 1 if (x == 'No' or x == 'Yes') else 0),
          #добавляет признак наличия интернета в пакете услуг клиента
          have_internet = lambda x: x['internet_service'].apply(lambda x: 1 if (x == 'Fiber optic' or x == 'DSL') else 0)
        ) 
      #заменяет строковые значения на числа для дополнительных услуг
      .pipe(change_on_number, columns_additional_service)
      #изменяет тип данных для дополнительных услуг
      .pipe(change_type, columns_additional_service, 'int')
      #считает количество услуг клиента
      .assign(count_of_service = lambda x: x.apply(count_of_service, axis=1)) 
)

df.head()

```

Все корректно. Проверим, корректно ли рассчитан показатель итоговой суммы выплат клиента. Для этого умножим количество месяцев, в течении которых действовал договор, на ежемесячный платеж.

```python
df['total'] = df['monthly_charges']*(df['month_active'])

display(df[['total_charges','total','monthly_charges', 'month_active','date_start','end_date']].head())

print('Среднее абсолютное отклонение расчетной итоговой суммы выплат клиента от фактической: {:.1f} д.е.'.format(
      mean_absolute_error(df['total_charges'],df['total'])))
    
```

Расхождение расчетной итоговой суммы выплат и фактической получилось значительной - 45.1 д.е. Возможно, это связано с учетной системой оператора. Например, по факту клиент заключил договор в середине декабря, а в учетной системе оператора начальная дата фиксируется как 01.12. В таком случае месяц использования для клиента считается с 15 по 15 число, а в системе учета оператора - с 1 по 1 число. 
Вторая возможная причина - некорректная выгрузка данных. 

Для построения модели будем использовать только сумму ежемесячного платежа, поэтому удалим итоговую сумму из признаков, так как они коррелируют.

Как уже ранее отметили, будем использовать для построения можели только сумму ежемесячного платежа (monthly_charges). Так как **количество услуг (count_of_service)** также сильно коррелирует с ежемесячным платежом, то его использовать для построения модели не будем. **Количество активных месяцев (month_active)** коррелирует с итоговой суммой, а вот с ежемесячным платежом нет, поэтому будем его использовать при построение модели. 

С целевым признаком сильной корреляции количественных признаков не выявлено.

```python
def plot_month_active(df, title):
    '''Строит гистограмму распределения количества активных месяцев по заданным значениям'''
    #строит гистограмму
    df.plot(kind='hist',
            y='month_active',
            grid=True,
            bins=30,
            figsize=(9,6))
    
    #выводит формат для гистограммы
    plt.title('Распределение количества месяцев использования услуг оператора в разрезе '+title+' клиентов')
    plt.xlabel('Количество месяцев, месяц')
    plt.ylabel('Количество договоров, штук')


#распределение по всем клиентам
plot_month_active(df, 'ВСЕХ')

#распределение по активным клиентам
plot_month_active(df.query('active_cust == 0'), 'АКТИВНЫХ')

#распределение по активным клиентам
plot_month_active(df.query('active_cust == 1'), 'УШЕДШИХ')

plt.show()
```

Видим, что на гистограмме распределения количества месяцев использования услуг оператора в разрезе всех клиентов присутствует два пика - в районе 0 и в районе 70 месяцев. ### Учет дисбаланса классов

В данных присутствует дисбаланс классов. Чтобы их сбалансировать используем три метода:
- Взвешивание
- Техника upsampling
- Техника downsampling

Для метода "взвешивание" не нужно преобразовывать обучающую выборку, в отличии от последних двух техник. Для них мы создадим новые обучающие выборки. 

#### Техника Upsampling

Еще раз посмотрим на соотношение классов в обучающей выборке.

```python
print(features_train.shape)
print(target_train.shape)
target_train.value_counts().to_frame()
```

Класс 1 (ушедшие клиенты) представлен меньшим количеством объектов.

```python
print('Увеличить класс 1 в {:.0f} раз'.format(3944/1495))
```

Чтобы сбалансировать классы, необходимо увеличить класс 1 в 3 раза. Напишем функцию, которая сбалансирует классы путем увеличения их числа.

```python
def upsample(features, target, repeat):
    '''Возвращает сбалансированные классы, скопировав несколько раз и перемешав объекты меньшего класса'''
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    return shuffle(features_upsampled, target_upsampled, random_state=13)
```

Сбалансируем классы обучающей выборки.

```python
features_train_upsampled, target_train_upsampled = upsample(features_train, target_train, 3)

print(features_train_upsampled.shape)
print(target_train_upsampled.shape)
target_train_upsampled.value_counts().to_frame()
```

Классы сбалансированы техникой Upsampling. Теперь сбалансируем техникой Downsampling.

#### Техника Downsampling

Согласно технике Downsampling необходимо уменьшить количество данных нулевого класса (активные клиенты). Рассчитаем, какую долю данных необходимо оставить.

```python
print('Оставить {:.0%} данных класса 0'.format(1495/3944))
```

Чтобы сбалансировать классы, необходимо оставить 38% от исходных данных нулевого класса. Напишем функцию, которая сбалансирует классы путем уменьшения их числа.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_train_downsampled, target_train_downsampled = downsample(features_train, target_train, 0.38)
```

Сбалансируем классы обучающей выборки.

```python
print(features_train_downsampled.shape)
print(target_train_downsampled.shape)
target_train_downsampled.value_counts().to_frame()
```

### Функции для машинного обучения

Напишем функции, которые будем использовать при построение моделей.

Для подбора гиперпараметров, обучения модели и определения необходимого на это времени напишем функцию `grid_fit`.

```python
def grid_fit(model, features, target, parameters):
    '''
    Принимает на входе модель, признаки и целевой признак и параметры для GridSearchCV. 
    Возвращает лучшую модель и время, потраченное на подбор гиперпараметров и обучение модели.
    ''' 
    start = time.time()
    
    #подбор гиперпараметров
    model = GridSearchCV(model, 
                         parameters, 
                         scoring='roc_auc', 
                         cv=shuffle, 
                         verbose=100,
                         n_jobs= -1)

    #обучение модели
    model.fit(features, target)
    
    end = time.time()

    #время обучения
    time_full = str(timedelta(seconds=(end-start)))
    
    print('Лучшая оценка: ', model.best_score_)
    print('Лучшие гиперпараметры модели: ', model.best_params_)
    print('Время обучения и подбора параметров:', time_full)
    
    return model.best_estimator_, time_full
                
```

Также напишем функцию, котрая будет рассчитывать метрики AUC-ROC, Recall, Precision и Accuracy, а также строить ROC-кривую.

```python
#создает функцию для оценки f1-меры, AUC_ROC и построения ROC-кривой
def auc_roc_accuracy_score(model,features,target):
    """
    Возвращает значения AUC-ROC, Recall, Precision и Accuracy, а также строит ROC-кривую для заданной модели
    """
    
    predictions = model.predict(features)    
    probabilities_valid = model.predict_proba(features)
    probabilities_one_valid = probabilities_valid[:, 1]
    
    auc_roc_score = roc_auc_score(target, probabilities_one_valid)
    accuracy_score = model.score(features, target)
    
    print('AUC-ROC: {:.3f}'.format(auc_roc_score))
    print('Accuracy: {:.3f}'.format(accuracy_score))
    print('Recall: {:.3f} '.format(recall_score(target, predictions)))
    print('Precision: {:.3f} '.format(precision_score(target, predictions)))
 
    #Построение ROC-кривая
    fpr, tpr, thresholds = roc_curve(target, probabilities_one_valid)
    plt.figure()

    #ROC-кривая модели
    plt.plot(fpr, tpr)

    #ROC-кривая случайной модели
    plt.plot([0, 1], [0, 1], linestyle='--')

    #задает форматирование
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show()
    
    return auc_roc_score, accuracy_score
```

Стоит отметить, хотя большинство клиентов с автоматической оплатой - это клиенты, заключившие договор с оператором на 2 года, часть клиенты с типом договора "от месяца к месяцу" тоже предпочитают автоматическую оплату, особенно в 2018-2019 гг.

6. Клиенты, подключившие интернет через оптоволоконный кабель (Fiber optic), уходият чаще, чем клиенты, использующие интернет через телефонную линию (DSL). Можно сделать вывод, что у оператора есть проблемы с интернетом через оптоволоконный кабель, это может быть связано с его качеством или же стоимостью.


7. В последние годы клиенты предпочитают одну телефонную линию. Доля клиентов, использующих несколько телефонных линий, снизилось с 73% в 2014 году до 26% в 2019 году. Клиенты с несколькими телефонными линиями уходят чаще, чем клиенты с одной телефонной линией.


8. Чаще уходят клиенты с более высокой стоимостью услуг. Средняя ежемесячная стоимость услуг активных клиентов - 61 д.е., а ушедших клиентов - 74 д.е., а медианы стоимости равны 64 и 80 д.е. соответственно.
При этом средняя ежемесячная стоимость услуг оператора снижается с каждым годом, при этом у ушедших покупателей она выше, в последние два года (2018-2019гг) разница почти в 50%. Также стоит отметить, что в первые годы существовала разница в стоимости по типу договора (на месяц, на год и на 2 года), долгосрочные договора были значительно выгоднее, но с каждым годом эта разница снижалась и в 2019 году она незначительная.


**Возможные причины ухода клиентов:**
1. Высокая стоимость услуг.
2. Низкое качество услуг, вызванное неготовностью инфраструктуры оператора к высоким нагрузкам.
3. Договор с оператором стало легче расторгнуть, так как преобладающее большинство клиентов заключают договор с помесячной оплатой вместо долгосрочных на 1-2 года.



**Выводы на основании построенных моделей:**

1. Все три модели дерева решений с использованием разных техник сбалансирования показали последний результат.


2. Модели случайного леса и CatBoost показали близкий результат. 


3. Наилучший результат у моделей RandomForestClassifier с использованием техники Downsampling и CatBoost со взвешиванием. Так как при изпользовании техники Downsampling обучающая выборка снижается почти в два раза, результат сильно зависит от разделения нулевого класса, поэтому приоритет отдаем модели CatBoost со взвешиванием.

4. Ключевой признак для модели является количество месяцев, в течении которых пользователь является клиентом оператора.  
Также значимые следующие признаки:
    - тип договора (оплата за месяц, год или 2 года),
    - тип интернета (через оптоволоконный кабель (Fiber optic) или через телефонную линию (DSL)),
    - ежемесячная стоимость услуг,
    - способ оплаты (автоплатеж или оплата по чеку).
