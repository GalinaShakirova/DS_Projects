## Аналитика

Построим распределение среднедневного количества прибывающих рейсов в разрезе городов за август 2018г.

```python
#задает размер гистограммы
plt.figure(figsize=(20,6))

#строит гистограмму
sns.barplot(x='city', 
            y='average_flights', 
            data=city.sort_values(by='average_flights'),
            palette='Blues_d')

#выводит формат для гистограммы
plt.title('Распределение среднего количества прибывающих рейсов в день в разрезе городов в август 2018г.',
          fontsize=18)
plt.xlabel('Город', fontsize=18)
plt.ylabel('Количество рейсов', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

Наибольшее среднедневное количество прибывающих рейсов в Москве. Данный показатель для Москвы в разы превышает показатели остальных городов России. 

Сформируем топ-10 городов с наибольшем количеством среднедневных прибывающих рейсов и построим график.

```python
city.sort_values(by='average_flights', ascending=False).head(10)
```

```python
#задает размер гистограммы
plt.figure(figsize=(15,6))

#строит гистограмму
sns.barplot(x='city', 
            y='average_flights', 
            data=city.sort_values(by='average_flights', ascending=False).head(10),
            palette='Blues_d')

#выводит формат для гистограммы
plt.title('Распределение среднего количества прибывающих рейсов в день в разрезе городов в август 2018г.',
          fontsize=15)
plt.xlabel('Город', fontsize=15)
plt.ylabel('Количество рейсов', fontsize=15)
plt.show()
```

В списке топ-10 городов по количеству рейсов в день за август 2018г на первом месте Москва (в среднем примерно 130 рейсов), затем Санкт-Петербург (31 рейс) и Новосибирск (17 рейсов).

Посмотрим на топ-10 городов с наименьшим количеством рейсов в день.

```python
city.sort_values(by='average_flights').head(10)
```

Во все 10 представленных городов в среднем в день прибывает лишь один рейс. Это, в большинстве своем, отдаленные территории страны.

Построим распределение количества рейсов в сентябре 2018 года в разрезе моделей самолетов.

```python
#задает размер гистограммы
plt.figure(figsize=(12,6))

#строит гистограмму
sns.barplot(x='model', 
            y='flights_amount', 
            data=model.sort_values(by='flights_amount'),
            palette='Blues_d')

#выводит формат для гистограммы
plt.title('Распределение количества рейсов в разрезе моделей самолетов в сентябре 2018г.', fontsize=15)
plt.xlabel('Модель', fontsize=15)
plt.ylabel('Количество рейсов', fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

Из гистограммы видим, что наибольшее количество рейсов в сентябре 2018 года было у моделей: 'Cessna 208 Caravan', 'Bombardier CRJ' и 'Sukhoi SuperJet-100'.

Согласно одной из классификации, а именно по дальности полета, самолеты можно отнести к одной из следующих групп:
- дальнемагистральные — свыше 6000 км;
- среднемагистральные — от 2500[5] до 6000 км;
- ближнемагистральные — от 1000 до 2500 км;
- самолёты местных воздушных линий — менее 1000 км.

Представленные самолеты компаний Boeing и Airbus относятся к средне и дальнемагистральным:
- Airbus A319-100 - 6700 км, 156 мест
- Airbus A321-200 - 5950 км, 220 мест
- Boeing 777-300 - 11 165 км, 451 мест
- Boeing 767-300 - 9700 км, 375 мест
- Boeing 737-300 - 4204 км, 149 мест

Самолет **Sukhoi SuperJet-100 (4578 км, 108 мест)** также относится к среднемагистральным, но максимальное количество мест у него меньше, чем у аналогичного Boeing 737-300.

Самолет **Bombardier CRJ-200 (3045 км, 50 мест)** также относится к среднемагистральным самолетам, но вот максимальное количество мест у него всего 50.

И только лидер по количеству рейсов за сентябрь 2018г самолет **Cessna 208 Caravan	(1 982 км, 14 мест)** относится к категории ближнемагистральных самолетов, у которого всего 14 мест. 

Из чего можем сделать вывод, что в сентябре более востребованы были перелеты на близкие и средние дистанции для небольшого количества пассажиров. А перелетов на дальние дистанции и с большим пассажиропотоком было значительно меньше.

А перелетов на дальние дистанции и с большим пассажиропотоком было значительно меньше.
