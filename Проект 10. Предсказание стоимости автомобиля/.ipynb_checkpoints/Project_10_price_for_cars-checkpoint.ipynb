# Предсказание стоимости автомобиля

**Заказчик**: сервис по продаже автомобилей с пробегом «Не бит, не крашен»

Сервис разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. 

**Входные данные**: исторические данные - технические характеристики, комплектации и цены автомобилей.

**Задача:** Вам нужно построить модель для определения стоимости. 

Заказчику важны:

1. качество предсказания;
2. скорость предсказания;
3. время обучения.


**Признаки:**

- DateCrawled — дата скачивания анкеты из базы
- VehicleType — тип автомобильного кузова
- RegistrationYear — год регистрации автомобиля
- Gearbox — тип коробки передач
- Power — мощность (л. с.)
- Model — модель автомобиля
- Kilometer — пробег (км)
- RegistrationMonth — месяц регистрации автомобиля
- FuelType — тип топлива
- Brand — марка автомобиля
- NotRepaired — была машина в ремонте или нет
- DateCreated — дата создания анкеты
- NumberOfPictures — количество фотографий автомобиля
- PostalCode — почтовый индекс владельца анкеты (пользователя)
- LastSeen — дата последней активности пользователя

**Целевой признак:** Price — цена (евро)

### Работа с пропущенными значениями

Перейдем к пропусакам. Начнем с наличия ремонта. 

```python
repair = df.groupby('not_repaired')['not_repaired'].count().to_frame()
repair['part'] = repair['not_repaired']/repair['not_repaired'].sum()
repair
```

Видим, что лишь 13% автомобилей были в ремонте. Проверим, есть ли взаимосвязь между наличием ремонта и пробегом и маркой автомобиля. Сформируем сводную таблицу.

```python
repair_kil = df.pivot_table(index=['kilometer'],
                            columns='not_repaired',
                            aggfunc={'registration_month':'count'})

#добавляет долю наличия ремонта
repair_kil['part_yes'] = repair_kil['registration_month', 'yes']/repair_kil.sum(axis=1)

repair_kil
```

Получили ожидаемый результат - чем больше пробег, тем больше доля автомобилей с ремонтом. Интересно, что наибольшая доля ремонта для новых автомобилей. Тут может быть несколько причин:
- Новые автомобилисты часто предпочитают новый автомобиль и по своей неопытности попадают в аварии.
- Денежный фактор - большинство автомобилей приобретаются в кредит вместе с КАСКО, следовательно ремонт происходит за счет страховой компании. В таком случае вероятность осуществления ремонта, если он необходим увеличивается.
- Психологический фактор - к новым автомобилям люди относятся бережнее, а значит при каких-то поломках, царапинах и др. более вероятно прибегнут к ремонту.
- Отзывные компании производителя, когда находят какой-либо брак при производстве.

Посмотрим на взаимосвязь марки автомобиля и ремонта.

```python
repair_brand = df.pivot_table(index='brand',
                              columns='not_repaired',
                              aggfunc={'registration_month':'count'})

#добавляет долю наличия ремонта
repair_brand['part_yes'] = repair_brand['registration_month', 'yes']/repair_brand.sum(axis=1)

repair_brand.sort_values(by='part_yes', ascending=False)
```

Чаще всего в ремонте оказываются автомобили марок: "Rover", "Daewoo" и "Lancia", а меньше всего - "Skoda", "Dacia", "Smart" и "Mini". 

На основании полученных данных приняли решения все пропуски заменить на "no", так как все-такие их преобладающее большинство.

```python
df['not_repaired'] = df['not_repaired'].fillna('no')
print('Количество пропусков:', df['not_repaired'].isna().sum())
```

Пропуски в параметре "Наличие ремонта" заменили. Перейдем к параметру **fuel_type** -  тип топлива. Сформируем сводную таблицу.

```python
fuel = df.groupby('fuel_type')['fuel_type'].count().to_frame()
fuel['part'] = fuel['fuel_type']/fuel['fuel_type'].sum()
fuel.sort_values(by='part', ascending=False)
```

Примерно 67% автомобилей имеют тип топлива - бензин, а почти 31% - газ. Посмотрим на взаимосвязь марки и типа топлива.

```python
fuel_brand = df.pivot_table(index='brand',
                            columns='fuel_type',
                            aggfunc={'registration_month':'count'})

#переименовывает параметры
fuel_brand.columns = ['cng','electric', 'gasoline', 'hybrid', 'lpg','other', 'petrol']

#добавляет долю типа топлива для бренда
fuel_brand['part_petrol'] = fuel_brand['petrol']/fuel_brand.sum(axis=1)
fuel_brand['part_gasoline'] = fuel_brand['gasoline']/fuel_brand.sum(axis=1)

#меняет местами параметры
new_order= ['part_petrol', 'part_gasoline', 'petrol', 'gasoline','cng','electric','hybrid','lpg','other']
fuel_brand = fuel_brand.reindex(new_order, axis=1)

fuel_brand.sort_values(by='part_petrol',ascending=False)
```

Преобладающее большинство марок автомобилей имеют тип топлива - бензин. Небольшая часть автомобильных брендов используют два типа топлива - бензин и газ - примерно в равных долях, такие бренды как  "Mercedes-Benz" и "Audi". Лишь для бренда "Land Rover" преобладает тип топлива - газ.

Для марки "Land_rover" заменим пропуски на газ, для всех остальных - на бензин. Для начала заменим все пропуски на 0.

```python
df['fuel_type'] = df['fuel_type'].fillna(0)
print('Количество пропусков:', df['not_repaired'].isna().sum())
```

Используя функцию, присвоим тип топлива.

```python
def fill_fuel_type(row):
    """Принимает строку DataFrame."""
    brand = row['brand']
    fuel_type = row['fuel_type']
    """
    Если в 'fuel_type' текущей таблицы значение 0 и бренд автомобиля "Land_rover, то возвращает 
    значение "gasoline", для остальных марок - "petrol". 
    Если нет - возвращает текущее значение параметра в исходной таблице.
    """
    if (brand == 'land_rover') and (fuel_type == 0):
        return 'gasoline'
    
    elif (fuel_type == 0):
        return 'petrol'
    
    return fuel_type

#заменяет пропуски
df['fuel_type'] = df.apply(fill_fuel_type, axis=1)
```

Проверим, как изменилось распределение данных.

```python
fuel = fuel.merge(df.groupby('fuel_type')['fuel_type'].count(), left_index=True,
    right_index=True,  suffixes=('_before', '_after'), how='left')

fuel['part_after'] = fuel['fuel_type_after']/fuel['fuel_type_after'].sum()

fuel.sort_values(by='part', ascending=False)
```

Заполнение пропусков не исказило данные, доли изменились незначительно. Перейдем к параметру **gearbox** - тип коробки передач. Сформируем сводную таблицу.

```python
gearbox = df.groupby('gearbox')['gearbox'].count().to_frame()
gearbox['part'] = gearbox['gearbox']/gearbox['gearbox'].sum()
gearbox
```

Почти 80% автомобилей имеют ручную коробку передач и лишь 20% - автоматическую. Посмотрим распределение по брендам.

```python
gearbox_brand = df.pivot_table(index='brand',
                               columns='gearbox',
                               aggfunc={'registration_month':'count'})

gearbox_brand.columns = ['auto', 'manual']

#добавляет долю типа коробки передач для бренда
for column in gearbox_brand.columns:
    name = 'part_'+column
    gearbox_brand[name] = gearbox_brand[column]/(gearbox_brand['auto'] + gearbox_brand['manual'])

gearbox_brand.sort_values(by='part_manual')
```

Есть бренды автомобилей, в которых преобладает автоматическая коробка передач - "Smart", "Jaguar" и "Jeep".
Есть бренды, в которых примерно одинаково используются оба типа коробки передач, например: "Chrysler", "Mercedes_benz", "Sonstige_autos" и "Land_rover". Но преобладающее большинство брендов предпочитают ручную коробку передач. 
Заполним пропуски следующим образом:
- для брендов, где преобладает автоматическая коробка передач - заполним пропуски на "avto",
- для тех, у кого примерно одинаково используются оба типа - рандомное значение, 
- для остальных - "manual".

Заменим пропуски на 0.

```python
df['gearbox'] = df['gearbox'].fillna(0)
print('Количество пропусков:', df['gearbox'].isna().sum())
```

Используя функцию, присвоим тип коробки передач.

```python
def fill_gearbox(row):
    """Принимает строку DataFrame."""
    brand = row['brand']
    gearbox = row['gearbox']
    """
    Если в 'fuel_type' текущей таблицы значение 0 и бренд автомобиля "Land_rover, то возвращает 
    значение "gasoline", для остальных марок - "petrol". 
    Если нет - возвращает текущее значение параметра в исходной таблице.
    """
    if (brand in ['smart','jaguar']) and (gearbox == 0):
        return 'auto'
    
    elif (brand in ['chrysler','mercedes_benz','sonstige_autos','porsche','land_rover']) and (gearbox == 0):
        return ''.join(pd.Series(['auto','manual']).sample(1))

    elif (gearbox == 0):
        return 'manual'
    
    return gearbox

#заменяет пропуски
df['gearbox'] = df.apply(fill_gearbox, axis=1)
```

Проверим, как изменилось распределение данных.

```python
gearbox_brand = gearbox_brand.merge(
    df.pivot_table(index='brand',columns='gearbox',aggfunc={'registration_month':'count'}),
    left_index=True,
    right_index=True, 
    how='left')

gearbox_brand.columns = ['auto','manual','part_auto','part_manual','auto_after',
       'manual_after']

#добавляет долю типа коробки передач для бренда
for column in ['auto_after','manual_after']:
    name = 'part_'+column
    gearbox_brand[name] = gearbox_brand[column]/(gearbox_brand['auto_after'] + gearbox_brand['manual_after'])
    
gearbox_brand.sort_values(by='part_manual')
```

Заполнение пропусков не исказило данные, доли изменились незначительно. Перейдем к параметру **vehicle_type** - тип автомобильного кузова. Сформируем сводную таблицу.

```python
vehicle = df.groupby('vehicle_type')['registration_year'].count().to_frame()
vehicle['part'] = vehicle['registration_year']/vehicle['registration_year'].sum()
vehicle.sort_values(by='registration_year',ascending=False)
```

```python
df.pivot_table(index=['brand','model'],
                   columns='vehicle_type',
                   aggfunc={'registration_month':'count'})
```

### Работа с некорректными данными

Для начала сформируем данные, на основании которых будем проводить проверку.
В качестве проверки заполнения мощности, значение которой ниже 20 л.с., возьмем бренд "Volkswagen" и модель "Golf".

```python
df_net.query('brand == "volkswagen" and model == "golf"').head()
```

Проверять замены высоких значений мощности будем также для бренда "Volkswagen" и модели "Golf" по строке с индексом **45216**.

```python
df_net.loc[45216,:].to_frame().T
```

Напишем функцию для замены некорректных значений.

```python
def fill_power(row, data):
    """Принимает строку и DataFrame."""
    brand = row['brand']
    model = row['model']
    power = row['power']
    
    """
    Если в 'power' текущей таблицы значение меньше 20, а также марка и модель есть в индексах
    таблицы data, то возвращает значение new_value из таблицы data. 
    Если нет - возвращает текущее значение параметра в исходной таблице.
    """
    if ((brand,model) in data.index) and ((power < 20) or (power > 500)):
        
        new_value = data.loc[(brand,model),'power']
        return new_value
    
    return power

#заменяет некорректные значения на медиану
df_net['power'] = df_net.apply(lambda row: fill_power(row, power_brand), axis=1)
```

### Линейная регрессия

Первой построим модель линейной регрессии.

```python
%%time
model_line = LinearRegression()
shuffle = KFold(n_splits=5, random_state=13, shuffle=True)
line_rmse = abs(cross_val_score(model_line, features_train, target_train, cv=shuffle,
                                 scoring='neg_root_mean_squared_error'))

print('Средний итоговый RMSE: {:.2f}'.format(line_rmse.mean()),
     '\n')
```

Для линейной регрессии через кросс-валидацию получили лучшее значение метрики RMSE равное 2520.26 евро. Довольно много. Обучим модель, предскажем стоимость и посчитаем метрику на валидационной и тестовой выборках.

```python
%%time

#обучение модели
start_line_train = time.time()

model_line.fit(features_train,target_train)

end_line_train = time.time()

time_line_train = str(timedelta(seconds=(end_line_train-start_line_train)))
```

Напишем функцию для получения предсказаний и метрики, а также времени предсказания. Применим ее на валидационной выборке.

```python
def pred_with_time(model,features, target):
    start = time.time()
    prediction = model.predict(features)

    end = time.time()
    
    time_pred = str(timedelta(seconds=(end-start)))

    rmse = mean_squared_error(target, prediction, squared=False)
    print('RMSE: {:.2f}'.format(rmse),
         '\n')
    
    return rmse, time_pred

rmse_line_valid, time_line_pred_valid = pred_with_time(model_line,features_valid, target_valid)
```

На валидационной выборке получили значение RMSE равное 2497.71 евро. Теперь получим метрику для тестовой выборки.

```python
%%time
rmse_line_test, time_line_pred_test = pred_with_time(model_line,features_test, target_test)
```

На тестовой выборке получили RMSE - 2509.95 евро. Сформируем итоговую таблицу по всем моделям.

```python
result = pd.DataFrame({"Model": ["Linear Regression"],
                       "RMSE valid": [rmse_line_valid], 
                       "RMSE test": [rmse_line_test],
                       "Time grid": ['-'],
                       "Time train": [time_line_train],
                       "Time prediction valid": [time_line_pred_valid],
                       "Time prediction test": [time_line_pred_test],
                       
                      })

result
```

Получим коэффициенты параметров и выведем для количественных признаков.

```python
coefficients = []
for i in range(features_train.shape[1]):
    coefficients.append([features_train.columns[i],model_line.coef_[i].round(4)])

coefficients = pd.DataFrame(coefficients,columns=['parametr','score'])
coefficients[coefficients['parametr'].isin(numeric_columns)].sort_values(by='score',ascending=False)
```
